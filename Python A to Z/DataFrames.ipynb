# Intro to Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
plt.rcParams["figure.figsize"]=8,4
```

```python
vis1 = sns.distplot(stats["InternetUsers"])
```

```python
vis1 = sns.distplot(stats["InternetUsers"], bins=20)
```

```python
vis2 = sns.boxplot(data=stats, x="IncomeGroup", y="BirthRate")
```

```python
vis3 = sns.lmplot(data=stats, x="InternetUsers", y="BirthRate", fit_reg=False, hue="IncomeGroup")
```

```python
vis3 = sns.lmplot(data=stats, x="InternetUsers", y="BirthRate", fit_reg=False, hue="IncomeGroup", height=8)
```

```python
vis3 = sns.lmplot(data=stats, x="InternetUsers", y="BirthRate", fit_reg=False, hue="IncomeGroup", height=8, scatter_kws={"s":200})
```
