```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
movies = pd.read_csv("P4-Section6-Homework-Dataset.csv", encoding="latin1")
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.columns
```

```python
movies.columns = ['DayOfWeek', 'Director', 'Genre', 'MovieTitle', 'ReleaseDate',
       'Studio', 'AdjustedGross($mill)', 'Budget($mill)', 'Gross($mill)',
       'IMDbRating', 'MovieLensRating', 'Overseas($mill)', 'Overseas%',
       'Profit($mill)', 'Profit%', 'Runtime(min)', 'US($mill)',
       'Gross%US']
```

```python
movies.info()
```

```python
movies.Genre = movies.Genre.astype("category")
movies.Studio = movies.Studio.astype("category")
movies.info()
```

```python
movies.Genre.cat.categories
```

```python
movies.Studio.cat.categories
```

```python
genre_filter = ["action","comedy","adventure","animation","drama"]
filter_genre_movies= movies[movies.Genre.isin(genre_filter)]
```

```python
studio_filter=["Buena Vista Studios","Sony", "Universal", "WB", "Paramount Pictures", "Fox"]
filter_movies= filter_genre_movies[filter_genre_movies.Studio.isin(studio_filter)]
```

```python
filter_movies.Genre.unique()
```

```python
filter_movies.Studio.unique()
```

```python
sns.set(style="darkgrid", palette="muted", color_codes=True)
ax = sns.boxplot(data=filter_movies, x="Genre", y="Gross%US", orient="v", color="lightgray", showfliers=False)
plt.setp(ax.artists, alpha=0.5)
sns.stripplot(x="Genre", y="Gross%US", data=filter_movies, jitter=True, size=6, linewidth=0, hue="Studio", alpha=0.7)

ax.axes.set_title("Domestic Gross % by Genre", fontsize=20)
ax.set_xlabel("Genre", fontsize=15)
ax.set_ylabel("Gross % US", fontsize=15)
ax.legend(bbox_to_anchor=(1.05,1), loc=2)
```
